```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
#plt.xkcd()
```

```python
trends = {"Harem Polygyny": -0.0112,
          "Monogamy": -0.0171,
          "Polygynandry": -0.0052, 
          "Polyandry": -0.0347,
          "Spatial Polygyny": -0.0174, 
          }

ya = 0

x = np.linspace(0, 75, 100)
ytrends = {regime: ya + x*b for regime, b in trends.items()}
```

```python
import pandas as pd

fig, ax = plt.subplots(figsize = (7,7))

ax.quiver(40,trends["Polyandry"]*55.5,0.18,0, scale = 1)
ax.quiver(40,trends["Polyandry"]*40,0,-0.19, scale = 1)
plt.annotate(r"$-0.0347$ myr$^{-1}$ = $\frac{dy}{dx}$", (2, -1.7), fontsize = 15)
plt.annotate(r"$y_a$", (-5,0))

colors = ["red", "orange", "blue", "green", "purple"]
for (regime, y), color in zip(ytrends.items(), colors):
    ax.plot(x, y, label = regime, color = color)

ax.set_ylabel("log ECV at 1 gram body mass")
ax.set_xlabel("time")
ax.set_xlim(-8, 80)

txt = r"""Caption: Brownian motion trend estimates per regime per time. The estimates (see table xx) and 
the plot is based on an arbitrary $y_a = 0$. Hence, the absolute trends give no information, but the 
contrasts in trends (e.g. Monogamy - Polyandry) give an account of relative rate of change per time."""

#plt.figtext(0.5, -0.06, txt, wrap=True, horizontalalignment='center', fontsize=12)
ax.legend(loc = [0.05, 0.05], framealpha = 1)

fig.show()

fig.savefig("trend_cartoon.pdf")
```
